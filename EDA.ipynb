```python
df['Total Household Income'] = df['Total Household Income'] / 1e3
df.describe()
```

### Distribution of household incomes

```python
g = sns.distplot(df['Total Household Income'], bins=30, kde=False, rug=False);
g.set(xlim=(0, 2000))

plt.xlabel('Total Household Income (Thousands of Pesos)')
plt.ylabel('Number of Households')
plt.title('Distribution of Total Household Income in Thousands of Pesos')
plt.show()

g.get_figure().savefig('income_histogram.png')
```

### How much does a Filipino family spend on food?

```python
g = sns.distplot(df['Total Food Expenditure'] / 1e3, kde=False, rug=False)
plt.xlabel('Yearly Total Food Expenditure in Thousands of Pesos')
plt.ylabel('Number of Families')
plt.title('Yearly Food Expenditure of Filipino Families')
plt.show()
g.get_figure().savefig('expenditure_histogram.png')
```

```python
cols = ['Bread and Cereals Expenditure', 'Total Rice Expenditure', 'Meat Expenditure', 
        'Total Fish and  marine products Expenditure', 'Fruit Expenditure', 'Vegetables Expenditure']

for col in cols:
    g = sns.distplot(df[col] / 1e5, kde=False, rug=False)
    g.set(xlim=(0, 1))
    plt.show()
```

```python
df = pd.read_csv('train.csv')
ax = sns.regplot(x="Total Food Expenditure", y="Total Household Income", data=df)
plt.xlabel('Yearly Total Food Expenditure (Pesos)')
plt.ylabel('Household Income (Pesos)')
plt.title('Effect of Yearly Total Food Expenditure on Household Income')
plt.show()
ax.get_figure().savefig('food_expenditure_income.png')
```

### How much does a Filipino Family Spend on Leisure?

```python
df = pd.read_csv('train.csv')
leisure = df['Restaurant and hotels Expenditure'] + df['Alcoholic Beverages Expenditure'] + \
        df['Tobacco Expenditure'] + df['Clothing, Footwear and Other Wear Expenditure'] + \
        df['Special Occasions Expenditure']
```

```python
ax = sns.regplot(x=leisure, y=df["Total Household Income"])
plt.xlabel('Yearly Total Leisure Expenditure (Pesos)')
plt.ylabel('Household Income (Pesos)')
plt.title('Effect of Yearly Total Leisure Expenditure on Household Income')
plt.show()
ax.get_figure().savefig('leisure_expenditure_income.png')
```

### How much does a family spend on housing?

```python
df = pd.read_csv('train.csv')
ax = sns.regplot(x="Housing and water Expenditure", y="Total Household Income", data=df)
plt.xlabel('Yearly Total Housing and water Expenditure (Pesos)')
plt.ylabel('Household Income (Pesos)')
plt.title('Effect of Housing and water Expenditure on Household Income')
plt.show()
ax.get_figure().savefig('Housing_and_water_Expenditure_income.png')
```

```python
numerics = []
categorical = []

for col in df:
    if((df[col].dtype == np.float64 or df[col].dtype == np.int64) and col != 'Unnamed: 0'):
        numerics.append(col)
        
    else:
        categorical.append(col)
        
categorical_df = df[categorical]
numeric_df = df[numerics]
```
