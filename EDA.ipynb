```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
print "Rows with missing values {0}".format(df.shape[0] - df.dropna().shape[0])
```

```python
print "Total rows {0}".format(len(df))
```

```python
df.describe()
```

```python
numerics = []
categorical = []

for col in df:
    if((df[col].dtype == np.float64 or df[col].dtype == np.int64) and col != 'Unnamed: 0'):
        numerics.append(col)
        
    else:
        categorical.append(col)
        
categorical_df = df[categorical]
numeric_df = df[numerics]
```
